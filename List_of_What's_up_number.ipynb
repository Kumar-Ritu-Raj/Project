```python
import pandas as pd
k=open('/content/text.txt')
p=set()
di=dict()
count=0
for i in k:
  i=i.rstrip()
  k=i.split()
  if '+91' in i:
    t=i[i.index('+91'):i.index('+91')+15]
    count+=1
    p.add(t)
    if t in di:
      di[t]=di[t]+1
    else:
      di[t]=1
        
```

```python
print("Your mobile Number =",p)
```

```python
print("Total Number of mobile number =",count)
c=0
l=open('/content/tex.xls','w+')
for aa in p:
  c+=1
  l.write(aa)
  l.write('\n')
l.close()
print("After removing duplicate mobile number= ",c)
print("Average chat per mobile no =",count/c)
```
