```python
#import required Library
from bs4 import BeautifulSoup
import requests
```

```python
# url for web scraping
url = 'https://www.javatpoint.com/'
```

```python
# access the url with request object
result = requests.get(url)

# scrape the webpage content
webpage = result.content

# create a soup object for passing web page using html parser
sl_soup = BeautifulSoup(webpage,'html.parser')
```

```python
# close the result object
result.close()
```

```python
# view the content of the soup object
sl_soup.contents
```

```python
# use prettify method to view well formated output (html tag)
print(sl_soup.prettify())
```

```python
# view the head of the soup object
print(sl_soup.head)
```

```python
# view the title of the web page 
sl_soup.title
```

```python
# find all the links present on web page
for href in sl_soup.findAll('a',href=True):
  print(href['href'])
```
